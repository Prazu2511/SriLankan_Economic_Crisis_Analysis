```python
data_t.columns = data_t.iloc[0].reset_index(drop=True)
data_t.head()
```

```python
data_t = data_t[1:]
```

```python
data_t.head()
```

```python
data_t= data_t.drop(columns=[
        'Agriculture-Agriculture Indices',
        'Central Government Finance-Central Government Revenue',
        'Central Government Finance-Debt Statistics',
        'Imports and Exports-Imports (in USD terms)',
        'Imports and Exports-Exports (in USD terms)',
        'Tourism-Tourist Arrivals'
    ])
```

```python
data_t.columns
```

```python
data_t.shape
```

```python
data_t.head()
```

```python
for col in data_t.columns:
    data_t[col].fillna(data_t[col].median(), inplace=True)
```

```python
data_t.head()
```

```python
data_t.tail()
```

```python
data_t.columns=[ 'Agriculture' , 'Revenue','Debts', 'Imports', 'Exports', 'Tourism']
```

```python
data_t.columns
```

```python
data_t.head()
```

```python
data_x = data_t.rename_axis("Year").reset_index()
```

```python
data_x.head()
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x="Year", y="Agriculture", data=data_x, marker='o')
plt.xticks(rotation=90)
plt.show
```

```python
plt.figure(figsize=(20, 10))

# Plot "Imports" and "Exports" against "Year"
sns.lineplot(x="Year", y="Imports", data=data_x, label="Imports", marker='o')
sns.lineplot(x="Year", y="Exports", data=data_x, label="Exports", marker='o')

# Set title and labels
plt.title("Imports and Exports Over Years")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend(title="Variables")
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
sns.pairplot(data_x[['Year', 'Imports', 'Exports', 'Tourism']])
plt.show()
```

```python

correlation_matrix = data_x.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
